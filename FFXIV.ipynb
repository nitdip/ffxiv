Pull all the items from the market board and assign each item name to an ID

```python
items = {}
pagetotal = market['Pagination']['PageTotal']
start = time.time()
for page in range(pagetotal):
    page += 1
    market = requests.get(f'https://xivapi.com/search?indexes=item&filters=ItemSearchCategory.ID>=9&page={page}').json()
    time.sleep(0.1)
    results = market['Pagination']['Results']
    for item in range(results):
        item_id = market['Results'][item]['ID']
        item_name = market['Results'][item]['Name']
        if item_name in items.keys():
            continue
        else:
            items[item_name] = item_id
print(f'took {time.time() - start} to run')
```

## Functions
We'll create a function that takes an item and sends a request to the API that brings back the historical prices for the item.

```python
# Searching for items on market
def xiv_request(server, item_name):
    '''
    Returns an item's price history from the market on a given server
    ----
    server: The server that your character is on
    item_name: The name of the item you're searching for
    ----
    returns list
    '''
    if item_name.title() not in item_dict.keys():
        raise ValueError(f'{item_name} not in Market Board')
    item_id = item_dict[item_name.title()]
    response = requests.get(f'https://xivapi.com/market/{server}/item/{item_id}').json()
    
    return response['History']
```
